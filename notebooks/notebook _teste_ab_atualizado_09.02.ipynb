## Hip√≥teses estat√≠sticas:
Antes de aplicar testes estat√≠sticos, √© fundamental garantir a qualidade e a confiabilidade dos dados por meio de uma an√°lise explorat√≥ria. Dados inconsistentes ou viesados podem levar a interpreta√ß√µes incorretas e comprometer a validade dos resultados do experimento.

### H0 (hip√≥tese nula):
N√£o h√° diferen√ßa estatisticamente significativa nas taxas de convers√£o entre os grupos controle e teste.

### H1 (hip√≥tese alternativa):
Os usu√°rios expostos ao novo sistema de recomenda√ß√µes apresentam taxas de convers√£o superiores √†s do grupo controle.

## Teste A/B


````python



from statsmodels.stats.proportion import proportions_ztest

def ztest_stage(success_B, total_B, success_A, total_A, stage_name):
    stat, p = proportions_ztest(
        count=[success_B, success_A],
        nobs=[total_B, total_A],
        alternative='larger'
    )
    print(f"\n TESTE Z ‚Äî {stage_name} (B > A)")
    print(f"Z = {stat:.4f} | p-value = {p:.6f}")
    if p < 0.05:
        print(" Evid√™ncia de que B > A")
    else:
        print(" Sem evid√™ncia de que B > A")

ztest_stage(397, 706, 1360, 2082, "Product Page")
ztest_stage(195, 706, 631, 2082, "Product Cart")
ztest_stage(198, 706, 652, 2082, "Purchase")



````

````python

# Etapas do funil
funnel_events = ['login', 'product_page', 'product_cart', 'purchase']

alpha = 0.05
test_name = 'recommender_system_test'

# --- usu√°rios do teste por grupo (dados limpos) ---
users_A = participants_clean.query(
    "ab_test == @test_name and group == 'A'"
)['user_id'].unique()

users_B = participants_clean.query(
    "ab_test == @test_name and group == 'B'"
)['user_id'].unique()

total_A = len(users_A)
total_B = len(users_B)

print(f"Teste: {test_name}")
print(f"Total usu√°rios A: {total_A}")
print(f"Total usu√°rios B: {total_B}")

# --- loop por etapa ---
for event in funnel_events:
    # usu√°rios que fizeram o evento em cada grupo
    event_A = events_test[
        (events_test['event_name'] == event) &
        (events_test['user_id'].isin(users_A))
    ]['user_id'].nunique()

    event_B = events_test[
        (events_test['event_name'] == event) &
        (events_test['user_id'].isin(users_B))
    ]['user_id'].nunique()

    # Teste unilateral B > A -> B deve estar primeiro
    count = np.array([event_B, event_A])
    nobs = np.array([total_B, total_A])

    z, p = proportions_ztest(count=count, nobs=nobs, alternative='larger')

    print(f"\nEvento: {event}")
    print(f"Convers√£o A: {event_A}/{total_A} ({event_A/total_A:.2%})")
    print(f"Convers√£o B: {event_B}/{total_B} ({event_B/total_B:.2%})")
    print(f"Z = {z:.4f} | p-value = {p:.6f}")

    if p < alpha:
        print("Resultado: evid√™ncia de que B > A")
    else:
        print("Resultado: sem evid√™ncia de que B > A")
````
